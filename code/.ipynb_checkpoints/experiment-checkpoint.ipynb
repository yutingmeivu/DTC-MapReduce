## DTCMR

load data

```python
cur_path = '/Users/yutingmei/Projects/DTC-MapReduce/data/'
d_s = pd.read_csv(cur_path + "trans_adult.csv")
```

format code for running DTCMR

```python
pool = mp.Pool(20)
precisionl = []
endl = []
subsample_num = 1000
group = 500
method = 'variance'
for i in range(10):
    precision, end = run_code(d_s, subsample_num, group, method)
    precisionl.append(precision)
    endl.append(end)
mpr = np.mean(precisionl)
mt = np.mean(endl)
pool.close()
print(f'mean computation time {round(mt, 3)} seconds with mean precision at {round(mpr, 3)} with total {subsample_num} samples with groups S = {group} with impurity function as {method}.')
```
